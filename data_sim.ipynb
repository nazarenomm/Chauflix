```python
movies_id = df_movies['id'].tolist()
```

```python
license_types = {1: 100_000, 2: 50_000, 3: 25_000}
durations = [30, 60, 90, 120, 180, 240, 300, 365]

licenses = []
license_payments = []

for id_license, id_movie in enumerate(movies_id, start=1):
    license_type = np.random.choice(list(license_types.keys()), p=[0.3, 0.5, 0.2])
    new_license = {
        'id': int(id_license),
        'movie_id': int(id_movie),
        'studio_id': None,
        'price': int(license_types[license_type] + np.random.randint(-10_000, 10_000)),
        'duration': int(np.random.choice(durations)),
        'type_id': int(license_type)
    }
    fecha = pd.Timestamp.now() - pd.Timedelta(days=new_license['duration']) + pd.Timedelta(days=np.random.randint(1, new_license['duration']))
    
    new_license_payment = {
        'license_id': int(id_license),
        'amount': int(new_license['price']),
        'date': fecha.strftime('%Y-%m-%d'),
    }

    licenses.append(new_license)
    license_payments.append(new_license_payment)
```

```python
insert_licenses = supabase_op.table('license').insert(licenses).execute()
insert_license_payments = supabase_op.table('license_payment').insert(license_payments).execute()

print(f"Licenses inserted: {insert_licenses.data}")
print(f"License payments inserted: {insert_license_payments.data}")
```
